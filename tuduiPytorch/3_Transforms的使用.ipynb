```python
from torchvision import transforms
from PIL import Image

#两个问题：1.如何使用transforms？2.为什么使用Tensor数据类型？
img_path = "dataset\\train\\ants_image\\0013035.jpg"
img = Image.open(img_path)
```

```python
tensor_trans = transforms.ToTensor()
tensor_img = tensor_trans(img)
```

```python
from torch.utils.tensorboard import SummaryWriter

writer = SummaryWriter("logss")
writer.add_image("Tensor_img", tensor_img)

writer.close()
```

### 常用的Transforms：

读取    PIL        Image.open()

转换   ToTensor   transforms.ToTensor()   转换为Tensor

标准化  Normalize  transforms.Normalize(mean, std)   归一化到[-1,1]

裁剪   Crop       transforms.CenterCrop(size)   裁剪中间部分

旋转   Rotate     transforms.RandomRotation(degrees)   随机旋转

翻转   Flipping   transforms.RandomHorizontalFlip()   随机水平翻转

缩放   Resize     transforms.Resize(size)   缩放到指定大小

色彩   Color      transforms.ColorJitter(brightness, contrast, saturation, hue)   随机调整色彩

锐化   Sharpness  transforms.RandomAdjustSharpness(sharpness_factor)   随机调整锐度
